```python
import pandas as pd

# 读取数据
file_path = r"E:\pythonProgram\Spillover\database\tweets\tweets.csv"
df = pd.read_csv(file_path)

# 转换 'time' 列为 datetime 格式，假设 'time' 列是字符串类型
df['time'] = pd.to_datetime(df['time'], format='%Y%m%d%H%M')
df = df[['time','id','text','isRetweet','isDeleted','device','favorites','retweets','isFlagged']]


# 定义日期范围
start_date_1 = pd.to_datetime('2017-01-20 00:00:00')
end_date_1 = pd.to_datetime('2021-01-20 00:00:00')

start_date_2 = pd.to_datetime('2021-01-20 00:00:00')
end_date_2 = pd.to_datetime('2025-01-20 00:00:00')

start_date_3 = pd.to_datetime('2025-01-20 00:00:00')
end_date_3 = pd.to_datetime('2025-03-14 12:00:00')

# 按照日期范围进行分割
df_first_term = df[(df['time'] >= start_date_1) & (df['time'] < end_date_1)]
df_interim_term = df[(df['time'] >= start_date_2) & (df['time'] < end_date_2)]
df_second_term = df[(df['time'] >= start_date_3) & (df['time'] <= end_date_3)]


df_first_term['time'] = df_first_term['time']
df_interim_term['time'] = df_interim_term['time']
df_second_term['time'] = df_second_term['time']
print(df_first_term.head())
print(df_interim_term.head())  
print(df_second_term.head())

# 保存为新的CSV文件
df_first_term.to_csv(r"E:\pythonProgram\Spillover\database\tweets\tweets_first_term.csv", index=False)
df_interim_term.to_csv(r"E:\pythonProgram\Spillover\database\tweets\tweets_interim_term.csv", index=False)
df_second_term.to_csv(r"E:\pythonProgram\Spillover\database\tweets\tweets_second_term.csv", index=False)

print("三个CSV文件已经成功保存。")
```

```python
import pandas as pd

# 读取CSV文件，指定分隔符为";"
df = pd.read_csv(r"D:\database\spillover\gc-1m.csv", header=None, delimiter=";")

# 合并前两个数据为time，格式为YYMMDDHHMM
df['time'] = pd.to_datetime(df[0].astype(str) + df[1].astype(str), format='%d/%m/%Y%H:%M:%S')

# 设置列名
df.columns = ['Date', 'Time', 'Open', 'High', 'Low', 'Close', 'Volume', 'time']

# 按时间排序
df = df.sort_values(by='time')

# 保留精确到分钟的时间并向上取整
df['time'] = df['time'].dt.ceil('T')

# 只保留需要的列
df = df[['time', 'Open', 'High', 'Low', 'Close', 'Volume']]

# 删除无用的Date和Time列
df.to_csv(r"D:\database\spillover\gc_new.csv", index=False)

print(df.head())
```

```python
import pandas as pd

# 从指定路径读取数据
df = pd.read_csv(r"D:\database\spillover\gc_new.csv")

# 确保time列是时间类型，并向上取整到最近的整分钟
df['time'] = pd.to_datetime(df['time'])

# 定义日期范围
start_date_1 = pd.to_datetime('2017-01-20 00:00:00')
end_date_1 = pd.to_datetime('2021-01-20 00:00:00')

start_date_2 = pd.to_datetime('2021-01-20 00:00:00')
end_date_2 = pd.to_datetime('2025-01-20 00:00:00')

start_date_3 = pd.to_datetime('2025-01-20 00:00:00')
end_date_3 = pd.to_datetime('2025-03-14 12:00:00')

# 筛选数据并保存到不同的文件

# 第一个任期
df_1 = df[(df['time'] >= start_date_1) & (df['time'] < end_date_1)]
df_1.to_csv(r"D:\database\spillover\gc_first_term.csv", index=False)

# 在野期
df_2 = df[(df['time'] >= start_date_2) & (df['time'] < end_date_2)]
df_2.to_csv(r"D:\database\spillover\gc_inactive_period.csv", index=False)

# 第二任期
df_3 = df[(df['time'] >= start_date_3) & (df['time'] <= end_date_3)]
df_3.to_csv(r"D:\database\spillover\gc_second_term.csv", index=False)

# 输出前三行检查
print(df_1.head())
print(df_2.head())
print(df_3.head())
```

```python
import pandas as pd

# 读取数据
file_path = r"E:\pythonProgram\Spillover\database\gold\gc_second_term.csv"
df = pd.read_csv(file_path)

# 确保 time 列为 datetime 类型
df['time'] = pd.to_datetime(df['time'])

# 按日期分组
df['date'] = df['time'].dt.date  # 提取日期

# 进行聚合
daily_df = df.groupby('date').agg(
    Open=('Open', 'first'),  # 取当天最早时间的 Open
    Close=('Close', 'last'),  # 取当天最晚时间的 Close
    High=('High', 'max'),  # 取当天最高值
    Low=('Low', 'min'),  # 取当天最低值
    Volume=('Volume', 'sum')  # 取当天总成交量
).reset_index()

# 保存日级数据
output_path = r"E:\pythonProgram\Spillover\database\gold\gc_daily.csv"
daily_df.to_csv(output_path, index=False)

# 显示前几行数据
print(daily_df.head())
```
